## **Slide 1: System Overview (MVT Architecture)**
**How Django bridges the gap between AI and User.**

Django follows the **Model-View-Template** architecture. In the BugsOut terminal:

* **Model (The Archive):** A database blueprint that stores 10 scientific features and the final AI prediction.
* **View (The Logic):** The "Brain" that cleans the raw input, runs the SVM algorithm, and decides if a particle is Gamma or Hadron.
* **Template (The Interface):** The Gold & Black high-visibility terminal that users use to interact with the system.



---

## **Slide 2: The Database (models.py)**
**Storing scientific telemetry data.**

Each particle entry consists of:
* **Scientific Measurements:** `fLength`, `fWidth`, `fSize`, etc., stored as `FloatField` for high precision.
* **Classification:** `prediction_label` stored as a `CharField` (Gamma or Hadron).
* **Traceability:** `created_at` records the exact second of the scan for the "Audit Log."

---

## **Slide 3: Machine Learning View (views.py)**
**Integrating the Support Vector Machine (SVM).**

When the user clicks **"INITIATE CLASSIFICATION"**, the view performs these steps:
1.  **POST Check:** Verifies the user submitted a valid form.
2.  **String Parsing:** Splitting the "bulk array" text into 10 separate numeric values.
3.  **SVM Hyperplane:** The model analyzes where the data sits relative to the decision boundary.
4.  **Database Commit:** Saving the result so it persists in the Audit History.

## **Slide 4: Advanced Features (URLs & Reporting)**
**Building professional-grade tools.**

* **Secure Routing:** Named URLs (e.g., `export_csv`) allow us to link buttons to Python logic without hardcoding addresses.
* **CSRF Security:** A digital handshake (`{% csrf_token %}`) that ensures only the authorized BugsOut terminal can talk to the server.
* **Data Portability:** An export function that streams database rows directly into a downloadable CSV file.
